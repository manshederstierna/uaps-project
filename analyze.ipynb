```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats


def confidenceInterval(data, confidence=0.95):
  mean = np.mean(data)
  std_err = stats.sem(data)  # Standard error of the mean
  return stats.t.interval(confidence=confidence, df=len(data)-1, loc=mean, scale=std_err)
```

```python
# Means of java and python fibonacci code
datafile = r"java_fib_times.txt"

data = pd.read_csv(datafile, delimiter=",")
n_values = [10,15,20,25,30,35,40] 
cutoff = 100

means = []
for n in  n_values:
    d = data[n == data["n"]]
    d = d.iloc[cutoff:]
    means.append(d["time"].mean())

plt.plot(n_values,means,marker="o",linestyle="-",label="Java")
plt.yscale("log")
plt.xlabel("n")
plt.ylabel("Time (ns)")


# Add python 
datafile = r"python_fib_times.txt"

data = pd.read_csv(datafile, delimiter=",")

n_values = [10,15,20,25,30,35,40] 
means_python = []
for n in  n_values:
    d = data[n == data["n"]]
    means_python.append(d["time"].mean())

plt.plot(n_values,means_python,marker="o",linestyle="-",label="Python")
plt.legend()
plt.show()
    
```

```python
# t-test java and python fibonacci

# Means of java and python fibonacci code
datafile = r"java_fib_times.txt"

data = pd.read_csv(datafile, delimiter=",")
n_values = [10,15,20,25,30,35,40] 
cutoff = 100


for n in  n_values:
    d = data[n == data["n"]]
    d = d.iloc[cutoff:]
    means.append(d["time"].mean())

# Add python 
datafile = r"python_fib_times.txt"

data = pd.read_csv(datafile, delimiter=",")

n_values = [10,15,20,25,30,35,40] 
means_python = []
for n in  n_values:
    d = data[n == data["n"]]
    means_python.append(d["time"].mean())

t_stat, p_value = stats.ttest_ind(np.log(means),np.log(means_python),equal_var=False)
print("p-värde logaritmerade medelvärden: ", p_value)
print("t-värde logaritmerade medelvärden: ",t_stat)

t_stat, p_value = stats.ttest_ind(means,means_python,equal_var=False)
print("p-värde medelvärden: ", p_value)
print("t-värde medelvärden: ",t_stat)




U1, p = stats.mannwhitneyu(means, means_python)
print("Mann-Whitneys U-test",p)


    
```

```python
# Konfidensintervall för respektive n
datafile = r"java_fib_times.txt"

data = pd.read_csv(datafile, delimiter=",")
n_values = [10,15,20,25,30,35,40] 
cutoff = 100


for n in  n_values:
    d = data[n == data["n"]]
    d = d.iloc[cutoff:]
    print(n," & ", round(d["time"].mean()/1000,3),"$\pm$", round((d["time"].mean()-confidenceInterval(d["time"])[0])/1000,3))
```

```python
#graph sort java
datafile = r"java_sort_times.txt"

data = pd.read_csv(datafile, delimiter=",")
data = data.iloc[150:5000]
plt.plot(range(150,data.shape[0]+150),data["time"],label="Java")

# graph sort 
datafile = r"python_sort_times.txt"

data_python = pd.read_csv(datafile, delimiter=",")
data_python = data_python.iloc[150:5000]


plt.plot(range(150,data_python.shape[0]+150),data_python["time"],label="Python")

plt.plot([150,5000],[data["time"].mean(),data["time"].mean()],linestyle="--",label="Java medelvärde",color="r")

plt.plot([150,5000],[data_python["time"].mean(),data_python["time"].mean()],linestyle="--",label="Python medelvärde")

plt.xlabel("Run")
plt.ylabel("Time (ns)")
plt.legend()
plt.show()

```

```python
# t-test sort java, python 

#graph sort java
datafile = r"java_sort_times.txt"

data = pd.read_csv(datafile, delimiter=",")
data_java = data.iloc[150:5000]

# graph sort 
datafile = r"python_sort_times.txt"

data = pd.read_csv(datafile, delimiter=",")
data_python = data.iloc[150:5000]

t_stat, p_value = stats.ttest_ind(data_java["time"],data_python["time"],equal_var=False)
print("p-värde: ", p_value)
print("t-värde: ",t_stat)

print("mean java time:", data_java["time"].mean())
print("python mean time:", data_python["time"].mean())

print("CI Java:", confidenceInterval(data_java["time"]))
print(data_java["time"].mean() - confidenceInterval(data_java["time"])[0])
print("CI Python:", confidenceInterval(data_python["time"]))
print(data_python["time"].mean() - confidenceInterval(data_python["time"])[0])
```
