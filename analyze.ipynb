```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

**Fibonacci java**

```python
datafile = r"C:\Users\clara\OneDrive\Dokument\LTH\Utvärdering_av_programvarusystem\uaps-project\java_fib_times.txt"

data = pd.read_csv(datafile, delimiter=",")

n_values = [10,15,20,25,30,35,40] 

for n in  n_values:
    d = data[n == data["n"]]
    x = range(d.shape[0])
    plt.plot(x,d["time"])
    plt.title(f"n={n}")
    plt.show()
```
